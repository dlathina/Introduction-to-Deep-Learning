## Simple Drawing

```python
# cropping 
img2=img[0:100, 0:100]
```

```python
cv2_imshow(img2)
```

```python
#sekarang kita ingin menggambar dengan kanvas img2
# gambar garis berwarna hijau

hijau = (0,255,0)
cv2.line(img2,(0,0),(25,50), hijau)
cv2_imshow(img2)
```

```python
# sekarang bagaimana cara menggambar kotak berwarna merah

merah =(0,0,255)
cv2.rectangle(img, (10,10),(75,25), merah)
cv2_imshow(img2)
```

```python
# buat ada fill di dalam kotak
# -1 pengertiannya adalah ketebalan maksimal

cv2.rectangle(img, (10,10),(75,25), merah, -1)
cv2_imshow(img2)
```

## Basic Image Processing

#### Gunakan matriks transformasi seperti di SMA dulu.

### Rotate gambar

#### Untuk rotation, pastikan titik centernya, dan pastikan sudut perputarannya

```python
# weight = columns
# height = rows
(h,w)=img2.shape[:2]
(cX, cY)=(w/2, h/2)
# M = cv2.getRotationMatrix2D(CENTER,ANGLE,SCALE ENLARGEMENT)
M = cv2.getRotationMatrix2D((cX,cY),45,1)
rotated = cv2.warpAffine(img2, M, (w,h))
cv2_imshow(rotated)
```
